Surface Defec Detetion Model

```python
from ultralytics import YOLO
```

```python
def train_surface_model():
    model = YOLO("yolov9c.pt")  
    model.train(
        data="defect.yaml",
        epochs=100,
        imgsz=640,
        batch=16,
        device=0,  # Use GPU
        project="surface_defect_model",
        name="yolov9_surface",
        pretrained=True,
        optimizer="SGD"
    )

    model.val()

if __name__ == "__main__":
    train_surface_model()
```
